```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
#import tqdm
# %matplotlib inline

images = glob.glob('./camera_cal/calibration*.jpg')
print (images)


objpoints = []
imgpoints = []

objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

for fname in images:
    print (fname)
    img=mpimg.imread(fname)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    ret, corners = cv2.findChessboardCorners(gray,(9,6), None)
    print ("Ret, corners",ret,corners)
    if ret == True:
    
        imgpoints.append(corners)
        objpoints.append(objp)
        
        
        img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
        zapis  = plt.figure(figsize=(20,20))
        
        last = fname.split("/")[2]
        file_name = "./camera_cal/new/" +last
        print (file_name)
        
        
        plt.imshow(img)
        zapis.savefig(file_name)
        plt.show()


ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)

np.save("objpoints.npy",objpoints)
np.save("imgpoints.npy", imgpoints)

print ("mtx",mtx)
print ("dist",mtx)

np.save("./camera_cal/new/mtx.npy",mtx)
np.save("./camera_cal/new/dist.npy", dist)

org_img = mpimg.imread('./test5.jpg')
plt.figure(figsize=(20,20))
plt.imshow(org_img)
plt.show()

dst = cv2.undistort(org_img, mtx, dist, None, mtx)

plt.figure(figsize=(20,20))
plt.imshow(dst)
plt.show()
```
